```python
import pandas as pd
import numpy as np
import seaborn as sns

pd.options.display.max_rows = None
pd.options.display.max_columns = None

datafile = "2022-train-v2.csv"
data = pd.read_csv(datafile)
```

```python
data.iloc[:, :6].plot(figsize=(14,7))
```

```python
print(data.columns[:].tolist())
```

```python
sns.heatmap(data.loc[:, 'clean_temp':'clean_pressure102'].corr())
```

```python
# 從數據集中選擇感興趣的特徵子集
interested_features = np.hstack((data[:, :6], data.loc[:, 'oven_pa1':'oven_b3']))

# 計算這個特徵子集之間的相關係數矩陣
corr_matrix = np.corrcoef(interested_features, rowvar=False)

# 繪製相關係數熱圖
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
```
